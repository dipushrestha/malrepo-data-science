# Modeling

Train and compare fraud detection models.

```python
import pandas as pd
import numpy as np
from src.models.trainer import ModelTrainer
from src.utils.config import load_config
```

```python
config = load_config('configs/model.yaml')
df = pd.read_parquet('data/processed/features.parquet')

target = 'is_fraud'
feature_cols = [c for c in df.select_dtypes(include=[np.number]).columns if c != target]
X = df[feature_cols].fillna(0)
y = df[target]
print(f'Features: {len(feature_cols)}, Samples: {len(X)}')
```

```python
# Train XGBoost
trainer = ModelTrainer(config)
metrics = trainer.train(X, y)
print(f'Metrics: {metrics}')
```

```python
# Cross validation
cv = trainer.cross_validate(X, y, cv=5)
print(f'CV F1: {cv["cv_f1_mean"]:.4f} +/- {cv["cv_f1_std"]:.4f}')
```

# Data Exploration

Exploratory data analysis of financial transactions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

df = pd.read_parquet('data/processed/features.parquet') if Path('data/processed/features.parquet').exists() else pd.DataFrame()
print(f'Shape: {df.shape}')
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Fraud distribution
if 'is_fraud' in df.columns:
    print(f'Fraud rate: {df["is_fraud"].mean():.4f}')
    df['is_fraud'].value_counts().plot.bar(title='Fraud Distribution')
    plt.show()
```

```python
# Transaction amount distribution
if 'transaction_amount' in df.columns:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    df['transaction_amount'].hist(bins=50, ax=axes[0])
    axes[0].set_title('Transaction Amount')
    np.log1p(df['transaction_amount']).hist(bins=50, ax=axes[1])
    axes[1].set_title('Log Transaction Amount')
    plt.tight_layout()
    plt.show()
```

```python
# Correlation matrix
numeric = df.select_dtypes(include=[np.number])
corr = numeric.corr()
plt.figure(figsize=(12, 10))
plt.imshow(corr, cmap='RdBu_r', aspect='auto')
plt.colorbar()
plt.title('Feature Correlations')
plt.show()
```
